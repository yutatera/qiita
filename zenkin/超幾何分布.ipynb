# 超幾何分布から二項分布

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom, hypergeom
```

###  M = 30, N=20のとき

```python
M = 30
N = 20
n = 10
p = N / M
plt.hist(hypergeom.rvs(M, n, N, size=100000), density=True, range=(-0.5, n + 0.5), bins=n+1)
plt.plot(range(n + 1), binom.pmf(range(n + 1), n, p), linewidth=4)
```
