# F分布からカイ二乗分布

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2
```

### 自由度(1, 1)

```python
n = 1
xmin = 0
xmax = 4
x = chi2.rvs(1, size=100000) / (chi2.rvs(n, size=100000) / n)
th = np.linspace(xmin, xmax, 100)
hi = np.array([np.sum(np.logical_and(th[i] < x, x <= th[i + 1])) for i in range(100 - 1)]) / (len(x) * (th[1] - th[0]))
plt.bar((th[:-1] + th[1:]) / 2, hi, width=(th[1] - th[0]))
s = np.linspace(xmin, xmax, 100)
plt.plot(s, chi2.pdf(s, 1), 'tab:orange', linewidth=4)
plt.xlim(xmin, xmax)
plt.ylim(0, 3)
```
