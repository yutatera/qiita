# F分布からカイ二乗分布

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2
```

### 自由度(1, 1)

```python
n = 1
xmin = 0
xmax = 4
x = chi2.rvs(1, size=100000) / (chi2.rvs(n, size=100000) / n)
x = x + (x > xmax) * (- x + xmax + (xmax - xmin) * 1 / 100)  # xmax以上のデータをxmax + (xmax - xmin) * 1 / 100にする。ビン数をプロット範囲内で約100にするため
plt.hist(x, bins=100, density=True)
s = np.linspace(xmin, xmax, 100)
plt.plot(s, chi2.pdf(s, 1), linewidth=4)
plt.xlim(xmin, xmax)
plt.ylim(0, 3)
```
