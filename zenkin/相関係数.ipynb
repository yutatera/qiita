# 相関係数

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
# y = x + epsilon.　x, epsilon ~ U(0,1)とする。相関係数の期待値は次のmu
mu = 1 / 2 * np.log((1 + 1 / 2 ** 0.5) / (1 - 1 / 2 ** 0.5))
```

### n = 5

```python
n = 5
rmin = -1
rmax = 3
rho = []
for _ in range(100000):
    x = np.random.rand(n)
    y = x + np.random.rand(n)
    r = np.corrcoef(x, y)[0, 1]
    rho.append(1 / 2 * np.log((1 + r) / (1 - r)))
plt.hist(rho, density=True, bins=200)
s = np.linspace(rmin, rmax, 100)
plt.plot(s, norm.pdf(s, mu, (1 / (n - 3)) ** 0.5), linewidth=4)
plt.xlim(rmin, rmax)
```
